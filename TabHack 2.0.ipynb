```python
X_train = df_train_final.drop('TARGET',axis = 1).values
y_train = df_train_final['TARGET']
```

```python
## Sampling the dataset
from imblearn.under_sampling import RandomUnderSampler
rus = RandomUnderSampler(sampling_strategy = 0.6,random_state=0)
X_resampled, y_resampled = rus.fit_resample(X_train, y_train)
```

```python
## Generating validation and training set
from sklearn.model_selection import train_test_split
X_train,X_valid,y_train,y_valid =  train_test_split( X_train, y_train, test_size=0.25, random_state=42)
```

```python
from sklearn.metrics import classification_report
```

#### Applying ML algorithms

```python
## Model 1
## Xgboost Classifier
import xgboost as xgb
xgb_cl = xgb.XGBClassifier()
```

```python
xgb_cl.fit(X_train, y_train)
```

```python
pred_xgb = xgb_cl.predict(X_valid)
print("---------------------------------------------------Classification Report---------------------------------------------")
print(classification_report(y_valid, pred_xgb))
```

#### Hyperparameter Tuning of XgBoost

```python
from sklearn.model_selection import RandomizedSearchCV, KFold
from xgboost import XGBClassifier
# XGBoost
# n_jobs=-1 to allow run it on all cores
params = {
    'n_estimators': [100, 200, 500],
    'learning_rate': [0.01,0.05,0.1,0.5],
    'booster': ['gbtree', 'gblinear'],
    'gamma': [0, 0.5, 1],
}

gs2 = RandomizedSearchCV(XGBClassifier(n_jobs=-1), params, n_jobs=-1, cv=KFold(n_splits=3), scoring='roc_auc',verbose = 10)
gs2.fit(X_resampled, y_resampled)
print('Best score:', gs2.best_score_)
print('Best score:', gs2.best_params_)
```

```python
## Catboost Classifier
from catboost import CatBoostClassifier
from sklearn.model_selection import GridSearchCV
```

```python
## Model 2
from catboost import CatBoostClassifier
cat_cl = CatBoostClassifier()
cat_cl.fit(X_train,y_train,verbose=False, plot=True)
```

```python
pred_cat = cat_cl.predict(X_valid)
print("---------------------------------------------------Classification Report---------------------------------------------")
print(classification_report(y_valid, pred_cat))
```

#### Hyperparameter Tuning of CatBoost Model

```python
cbc = CatBoostClassifier()

#create the grid
grid = {'max_depth': [4,6,8],'n_estimators':[100, 200, 300]}

#Instantiate GridSearchCV
gscv = GridSearchCV (estimator = cbc, param_grid = grid, scoring ='accuracy', cv = 5)

#fit the model
gscv.fit(X_resampled,y_resampled)

#returns the estimator with the best performance
print(gscv.best_estimator_)

#returns the best score
print(gscv.best_score_)

#returns the best parameters
print(gscv.best_params_)
```

```python
## Model 3
from sklearn.ensemble import AdaBoostClassifier
ada_cl = AdaBoostClassifier()
ada_cl.fit(X_train, y_train)
```

```python
pred_ada = ada_cl.predict(X_valid)
print("---------------------------------------------------Classification Report---------------------------------------------")
print(classification_report(y_valid, pred_ada))
```

```python
ab_clf = AdaBoostClassifier()
from sklearn.model_selection import RandomizedSearchCV
#create the grid
parameters = {
    'n_estimators': [2, 5, 6, 9, 10, 11, 12, 20],
    'learning_rate': [(0.97 + x / 100) for x in range(0, 5)],
    'algorithm': ['SAMME', 'SAMME.R']
}
clf = RandomizedSearchCV(ab_clf, parameters, cv=5, scoring ='accuracy')
clf.fit(X_resampled, y_resampled)
print(clf.best_estimator_)

#returns the best score
print(clf.best_score_)

#returns the best parameters
print(clf.best_params_)
```

#### Model Stacking

```python
classifier1 = xgb.XGBClassifier(n_estimators = 200, learning_rate = 0.1, gamma = 1, booster = 'gblinear')
classifier2 = CatBoostClassifier(max_depth = 4, n_estimators = 300)
classifier3 = AdaBoostClassifier(n_estimators = 10, learning_rate = 1.01, algorithm = 'SAMME.R')
```

```python
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
```

```python
estimators = [('xgb', classifier1),('cat',classifier2),('ada', classifier3)]
```

```python
clf = StackingClassifier( estimators=estimators, final_estimator=LogisticRegression())
```

```python
clf.fit(X_resampled,y_resampled)
```

```python
pred_stack = clf.predict(X_valid)
```

```python
print("---------------------------------------------------Classification Report---------------------------------------------")
print(classification_report(y_valid, pred_stack))
```
